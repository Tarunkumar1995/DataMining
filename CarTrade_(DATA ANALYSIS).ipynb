```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## 2. Which Company has the most number of sales ?
## Univariate Analysis


```python
plt.figure(figsize=(15,7))
sns.set(style='darkgrid')
chart=sns.countplot(x="Company",data=df,palette='Set1')
chart.set_xlabel("Company",fontsize=18)
chart.set_xticklabels(chart.get_xticklabels(),rotation=90)
plt.show()
```

###  Answer
- Maruti Suzuki has the most number of sold cars in the Country.
- and later comes Hyundai,and these two companies have more sold cars in which  no company was close to these companies.

# 3. Are Diesel or Petrol Cars sold the most?

```python
plt.figure(figsize=(15,7))
chart=sns.countplot(x='Fuel_Type',data=df,palette='Set2')
chart.set_xlabel("Fuel Type",fontsize=18)
plt.show()
```

## Answer
 - Diesel cars have been sold most which is more than 5000,and then followed by petrol.

# 4.Which city has most sold cars?

```python
plt.figure(figsize=(17,8))
chart=sns.countplot(x="Location",data=df)
chart.set_xlabel("Location",fontsize=20)
plt.show()
```

## Answer 
- Kolkata has the most number of sold cars.
- Ahemdabad,Banglore,New Delhi,gurgoan,Hyderabad,pune has the cars which was sold 1100(approx)
- Navi Mumbai has the least sold cars

## Which year the cars have been manufactured more?
